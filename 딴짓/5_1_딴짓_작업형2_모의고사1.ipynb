````python
import pandas as pd
#베이스라인
train=pd.read_csv("/content/drive/MyDrive/데이터모음/딴짓/ceredit card/train.csv")
test=pd.read_csv("/content/drive/MyDrive/데이터모음/딴짓/ceredit card/test.csv")
print(train.shape, test.shape)
train=train.select_dtypes(exclude='object')
test=test.select_dtypes(exclude='object')
print(train.shape, test.shape)
````

````python
#CLIENTNUM 삭제
train=train.drop('CLIENTNUM', axis=1)
testid=test.pop('CLIENTNUM')
print(train.shape, test.shape)
````

## 검증 데이터 분리

````python
from sklearn.model_selection import train_test_split

# xtr, xval, ytr, yval = train_test_split(train, target, test_size, random_state)
xtr, xval, ytr, yval = train_test_split(train.drop('Attrition_Flag', axis=1), train['Attrition_Flag'], test_size=0.1, random_state=2022)
````
